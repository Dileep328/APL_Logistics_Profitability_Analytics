# LOAD DATA

```python
import pandas as pd

df = pd.read_csv(
    "../data/processed/cleaned_data.csv"
)
```

# Executive KPIs

```python
total_revenue = df["Sales"].sum()

total_profit = df[
    "Order Profit Per Order"
].sum()

profit_margin = (
    total_profit /
    total_revenue
) * 100

total_customers = df[
    "Customer Id"
].nunique()

total_products = df[
    "Product Name"
].nunique()

total_markets = df[
    "Market"
].nunique()
```

# KPI Table

```python
executive_summary = pd.DataFrame({

    "KPI":[
        "Total Revenue",
        "Total Profit",
        "Profit Margin",
        "Customers",
        "Products",
        "Markets"
    ],

    "Value":[
        total_revenue,
        total_profit,
        profit_margin,
        total_customers,
        total_products,
        total_markets
    ]
})

executive_summary
```

# CUSTOMER INSIGHT

# Top Customers

```python
top_customers = (

    df.groupby("Customer Id")
    ["Order Profit Per Order"]

    .sum()

    .sort_values(
        ascending=False
    )

    .head(10)
)

top_customers
```

# Profit Contribution

```python
customer_profit = (

    df.groupby("Customer Id")
    ["Order Profit Per Order"]

    .sum()

    .sort_values(
        ascending=False
    )
)
```

```python
top20_profit_share = (

    customer_profit.head(
        int(
            len(customer_profit)*0.2
        )
    ).sum()

    /

    customer_profit.sum()

) * 100

top20_profit_share
```

Top 20% customers contribute the majority of profits.

Business should prioritize customer retention programs for these customers.

# Product Insights 

```python
category_profit = (

    df.groupby("Category Name")

    ["Order Profit Per Order"]

    .sum()

    .sort_values(
        ascending=False
    )
)

category_profit.head(10)
```

# Best Category

# Worst Category

# Discount Insights 

```python
discount_analysis = (

    df.groupby(
        "Discount_Bucket"
    )

    ["Profit_Margin"]

    .mean()
)
```

```python
discount_analysis
```

Profit margins decline consistently as discount rates increase.

Discounts above 20% should require approval.

# Market Insighs

```python
market_profit = (

    df.groupby("Market")

    ["Order Profit Per Order"]

    .sum()

    .sort_values(
        ascending=False
    )
)

market_profit
```

Europe and LATAM are major profit contributors.

Expansion strategies should prioritize these markets.
